```python
#Este pequeÃ±o bloque nos dice cuantas veces gano el equipo 1 y cuantas veces gano el equipo 2

ganador1=0
ganador2=0

for i in lol.winner:
    if i == 1:
     
        ganador1+=1
    else:
        
        ganador2+=1

print("Ganador 1= ", ganador1)
print("Ganador 2= ", ganador2)
```

```python
#Este pequeÃ±o bloque nos dice cuantas veces gano el equipo 1 tiro el primer inhbidor y cuantas veces el equipo 2 tiro el primer inhibidor

firstInhibitor1=0
firstInhibitor2=0

for i in lol.firstInhibitor:
    if i == 1:
     
        firstInhibitor1+=1
    else:
        
        firstInhibitor2+=1

print("Ganador 1= ", firstInhibitor1)
print("Ganador 2= ",  firstInhibitor2)
```

```python
#para acceder a mas de uno es necesario utilizar [[]]

lol[["winner", "firstInhibitor"]]

```

```python
#se crea el filtro
filter1 = lol.gameDurationMin > 50

#nuestro dataframe lol es filtrado y puesto en otro nombre creando un segundo dataframe
new_data = lol[filter1]

#este codigo nos devuelve un dataframe con todas las filas y la lista de columnas
new_data = new_data.loc[:,["winner","gameDurationMin","t1_towerKills","t1_inhibitorKills","t1_baronKills","t1_dragonKills","t2_towerKills",
            "t2_inhibitorKills","t2_baronKills","t2_dragonKills"]]

#se crea nueva variable que suma las valores de ambos equipos
inhip_total = new_data.t1_inhibitorKills + new_data.t2_inhibitorKills
#se la agrega al dataframe
new_data["Total_inhibitor"] = inhip_total

tower_total = new_data.t1_towerKills + new_data.t2_towerKills
new_data["Total tower"] = tower_total

baron_total = new_data.t1_baronKills + new_data.t2_baronKills
new_data["Total baron"] = baron_total

dragon_total = new_data.t1_dragonKills + new_data.t2_dragonKills
new_data["Total dragon"] = dragon_total


#printeamos nuestro nuevo dataframe
new_data
```

```python
#creacion nuevo filtro
filter2=new_data["Total dragon"]>5
```

```python
#creacion nuevo dataframe, con el dataframe anterior con el filtro aplicado
new_data_2=new_data[filter2]

#al nuevo dataframe creado en el paso anterior, recortamos las filas a las de abajo
new_data_2=new_data_2.loc[:, ["winner", "Total dragon", "Total baron"]]
```
